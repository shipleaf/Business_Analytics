```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
columns_to_analyze = ['Age', 'DailyRate', 'EnvironmentSatisfaction', 'Attrition']
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.scatterplot(x='Age', y='Attrition', data=data, ax=axes[0])
axes[0].set_title('Age vs Attrition')

sns.scatterplot(x='DailyRate', y='Attrition', data=data, ax=axes[1])
axes[1].set_title('DailyRate vs Attrition')

sns.scatterplot(x='EnvironmentSatisfaction', y='Attrition', data=data, ax=axes[2])
axes[2].set_title('EnvironmentSatisfaction vs Attrition')

plt.show()
```

```python
columns_to_analyze = ['Education', 'JobRole', 'JobLevel', 'JobSatisfaction']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.barplot(x='Education', y='JobSatisfaction', data=data, ax=axes[0])
axes[0].set_title('Education vs JobSatisfaction')

sns.barplot(x='JobRole', y='JobSatisfaction', data=data, ax=axes[1])
axes[1].set_title('JobRole vs JobSatisfaction')
axes[1].tick_params(axis='x', rotation=90)

sns.barplot(x='JobLevel', y='JobSatisfaction', data=data, ax=axes[2])
axes[2].set_title('JobLevel vs JobSatisfaction')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.countplot(x='WorkLifeBalance', hue='OverTime', data=data, ax=axes[0])
axes[0].set_title('WorkLifeBalance vs OverTime')

sns.barplot(x='WorkLifeBalance', y='PercentSalaryHike', hue='OverTime', data=data, ax=axes[1])
axes[1].set_title('WorkLifeBalance vs PercentSalaryHike')

plt.show()
```
